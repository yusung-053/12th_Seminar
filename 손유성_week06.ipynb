# Principal Component Analysis

```python
import numpy as np
import pandas as pd

# 공분산 행렬 정의
Sigma=np.array([[5,4,1,0],[4,6,4,1],[1,4,6,4],[0,1,4,5]])

# 고유값 분해 및 내림차순 정렬
eigvals,eigvecs=np.linalg.eig(Sigma)
idx=np.argsort(eigvals)[::-1]
eigvals=eigvals[idx]
eigvecs=eigvecs[:,idx]

# 각 주성분의 분산 비율
explained_var_ratio=eigvals/np.sum(eigvals)

# 결과 정리
result = pd.DataFrame({
    'A': eigvals,
    'B': explained_var_ratio,
}, index=[f'PC{i+1}' for i in range(len(eigvals))])

print(result)
print(pd.DataFrame(eigvecs, columns=[f'PC{i+1}' for i in range(4)], index=['x1', 'x2', 'x3', 'x4']))
```

# Scree Plot

```python
# Scree Plot 그리기
import matplotlib.pyplot as plt
plt.figure(figsize=(6, 4))
plt.plot(range(1, len(eigvals)+1), eigvals, 'o-', linewidth=2)
plt.title('Scree Plot')
plt.xlabel('Principal Component')
plt.ylabel('Eigenvalue (Variance)')
plt.xticks(range(1, len(eigvals)+1))
plt.grid(True)
plt.show()
```
